```sql
%python
import pyspark.sql.functions as F
# from pyspark.sql.types import IntegerType
df = spark.read.table("bronze_workforce.azure_blob_storage.payroll")
display(df)
```

```sql
%python
df=df.drop('_file','_line','_modified','_fivetran_synced')
```

```sql
%python
df = df.withColumn('pay_period_start',F.trim(F.col("pay_period_start")))
df = df.withColumn("pay_period_start", F.to_timestamp(F.col("pay_period_start"), "dd-MM-yyyy HH:mm"))
df = df.withColumn("pay_period_start", F.to_date(F.col("pay_period_start"), "dd-MM-yyyy"))
display(df)
```

```sql
%python
df = df.withColumn('pay_period_end',F.trim(F.col("pay_period_end")))
df = df.withColumn("pay_period_end", F.to_timestamp(F.col("pay_period_end"), "dd-MM-yyyy HH:mm"))
df = df.withColumn("pay_period_end", F.to_date(F.col("pay_period_end"), "dd-MM-yyyy"))
display(df)
```

```sql
%python
df = df.withColumn('pay_date',F.trim(F.col("pay_date")))
df = df.withColumn("pay_date", F.to_timestamp(F.col("pay_date"), "dd-MM-yyyy HH:mm"))
df = df.withColumn("pay_date", F.to_date(F.col("pay_date"), "dd-MM-yyyy"))
display(df)
```

```sql
 %python
df = df.dropDuplicates()
display(df)
```

```sql
%python
df = df.withColumn("payroll_id",df["payroll_id"].cast('int')).withColumn("employee_id",df["employee_id"].cast('int')).withColumn("company_id",df["company_id"].cast('int')).withColumn("department_id",df["department_id"].cast('int'))
display(df)
```

```sql
%python
df.write.format("delta").mode("overwrite").saveAsTable("silver_workforce.azure_blob_storage.payroll")
display(spark.sql("select * from silver_workforce.azure_blob_storage.payroll"))
```
