```python
from pyspark.sql.functions import monotonically_increasing_id

dim_employee = spark.table("silver_workforce.azure_blob_storage.employee") \
    .withColumn("employee_key", monotonically_increasing_id())


 
```

```python

dim_employee.write.format("delta").mode("overwrite").option("mergeSchema", "true").saveAsTable("gold_workforce.azure_blob_storage.dim_employee")
display(dim_employee)
```

```python

dim_accounts = spark.table("silver_workforce.azure_blob_storage.accounts") \
    .withColumn("account_key", monotonically_increasing_id())
display(dim_accounts)
```

```python
dim_accounts.write.format("delta").mode("overwrite").saveAsTable("gold_workforce.azure_blob_storage.dim_accounts")
```

```python

 
dim_company = spark.table("silver_workforce.azure_blob_storage.company") \
    .withColumn("company_key", monotonically_increasing_id())
 
dim_company = dim_company.select(
    'company_id','company_name','industry','country','company_key'
)
display(dim_company)
```
